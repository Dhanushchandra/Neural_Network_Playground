```python
from sentence_transformers import SentenceTransformer, util
```

```python
import torch
```

```python
documents = [
    "Apples and bananas are fruits",
    "I love driving my car on highways",
    "Fruits are rich in vitamins",
    "Cars need fuel to run",
    "Bananas are high in potassium"
]
```

```python
# Load model
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
doc_embeddings = model.encode(documents, convert_to_tensor=True)
```

```python
def semantic_search(query, top_k=4):
    query_embedding = model.encode(query, convert_to_tensor=True)
    scores = util.cos_sim(query_embedding, doc_embeddings)[0]

    top_results = torch.topk(scores, k=top_k)

    for score, idx in zip(top_results.values, top_results.indices):
        print(f"{documents[idx]}  --> score: {score:.4f}")
```

```python
semantic_search("food",3)
```
