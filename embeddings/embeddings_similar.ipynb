```python
import matplotlib.pyplot as plt
from sentence_transformers import SentenceTransformer
from sklearn.manifold import TSNE
```

```python
words = []
sentences = [
    "I love driving my car",
    "This automobile is very fast",
    "Cars are amazing machines",
    "Bananas are rich in potassium",
    "I ate a banana today",
    "Oranges and bananas are fruits"
]
```

```python
for sentence in sentences:
    for w in sentence.split():
        words.append(w)
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python

embeddings = model.encode(words)
```

```python
for word in words:
    print(word)
```

```python
# reduce to 2D
tsne = TSNE(n_components=2, perplexity=3, random_state=42)
points_2d = tsne.fit_transform(embeddings)
```

```python
plt.figure(figsize=(7,5))
for i, word in enumerate(words):
    plt.scatter(points_2d[i,0], points_2d[i,1])
    plt.text(points_2d[i,0]+0.2, points_2d[i,1]+0.2, word, fontsize=9)

plt.title("2D Visualization of Word Embeddings")
plt.show()
```
