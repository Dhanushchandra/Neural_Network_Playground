```python
import math
```

Minimal Log Loss function

`L(y,y^​)=−[ylog(y^​)+(1−y)log(1−y^​)]​`

```python
def log_loss(y, y_hat):
    """
    Binary log loss (cross-entropy)
    y     : true label (0 or 1)
    y_hat : predicted probability (0 < y_hat < 1)
    """
    # Small epsilon to avoid log(0) (optional)
    epsilon = 1e-15
    y_hat = max(min(y_hat, 1 - epsilon),epsilon)

    return - (y * math.log(y_hat) + (1 - y) * math.log(1 - y_hat))
```

Try it with an example

```python
#True label
y = 1

#Model prediction from forward pass
y_hat = 0.46

loss = log_loss(y,y_hat)

print("Prediction (ŷ):", y_hat)
print("True label (y):", y)
print("Log Loss:", round(loss, 4))
```

Different predictions when y = 1 (build intuition)

```python
for y_hat in [0.9, 0.7, 0.5, 0.3, 0.1]:
    loss = log_loss(1,y_hat)
    print(f"ŷ={y_hat} → loss={round(loss, 3)}")
```

when y = 0

```python
for y_hat in [0.9, 0.7, 0.5, 0.3, 0.1]:
    loss = log_loss(0, y_hat)
    print(f"ŷ={y_hat} → loss={round(loss, 3)}")
```
