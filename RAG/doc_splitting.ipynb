```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
text = """
# AI Implementation Guide
Version: 2025.1

## Introduction
Artificial Intelligence is transforming industries. It requires high-quality data.
If the data is poor, the model output will be poor.

## Technical Setup
To install the library, use the following command:
pip install langchain-text-splitters

### Code Example
import os
def start_app():
    print("App is starting...")
    # Initialize the engine
    engine = create_engine()
"""
```

```python
# 2. Customizing Separators
# We add "#" to respect Markdown headings
custom_separators = ["\n# ", "\n## ", "\n### ", "\n\n", "\n", " ", ""]
```

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size = 100,
    chunk_overlap = 20,
    separators = custom_separators
)
```

```python
chunks = splitter.split_text(text)
```

```python
for i, chunk in enumerate(chunks):
    print(f"--- Chunk {i+1} ({len(chunk)} chars) ---")
    print(chunk)
    print("-" * 30)
```
