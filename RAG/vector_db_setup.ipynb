```python
import chromadb
```

```python
client = chromadb.PersistentClient(path="my_vectordb")
```

```python
ls
```

```python
collection = client.get_or_create_collection(name="office_docs")
```

Add data
Chroma can handle the embedding for you if you don't want to do it manually!

```python
collection.add(
    documents=["Our office is in Bangalore.", "CEO is Sarah Jenkins."],
    metadatas=[{"type": "location"}, {"type": "person"}],
    ids=["id1", "id2"]
)
```

```python
# Fetch everything in the database
data = collection.get(include=['documents', 'metadatas', 'embeddings'])

for i in range(len(data['ids'])):
    print(f"ID: {data['ids'][i]}")
    print(f"Metadata: {data['metadatas'][i]}")
    # The first 5 numbers of the high-dimensional vector
    print(f"Vector (Snippet): {data['embeddings'][i][:5]}...") 
    print(f"Text: {data['documents'][i][:100]}...")
    print("-" * 20)
```

Search

```python
results = collection.query(
    # query_texts = ["Who is in charge?"],
    query_texts = ["Which building?"],
    n_results=1
)
```
