The NN learning loop

1. Forward pass  → get prediction
2. Compute loss  → how wrong was it?
3. Backpropagate → which weights caused the error?
4. Update weights → reduce future error
5. Repeat

Forward pass is the process of passing input data through a neural network layer by layer to compute the predicted output (ŷ).

```python
import math
```

```python
def sigmoid(z):
    return 1 / (1 + math.exp(-z))
```

```python
X = [
    (0,0),
    (0,1),
    (1,0),
    (1,1)
]
```

```python
def hidden_layer(x1, x2):
    # Neuron 1: OR-like
    z1 = 2*x1 + 2*x2 - 1
    a1 = sigmoid(z1)

    # Neuron 2: AND-like
    z2 = 2*x1 + 2*x2 - 3
    a2 = sigmoid(z2)

    return a1, a2
```

```python
def output_neuron(a1, a2):
    z = 3*a1 - 5*a2 - 1
    return sigmoid(z)
```

```python
print("X1 X2 => prediction")

for x1, x2 in X:
    a1, a2 = hidden_layer(x1, x2)
    y_hat = output_neuron(a1, a2)
    print(x1, x2, "→", round(y_hat, 2))
```
