````python
import math
````

````python
def sigmoid(z):
    return 1/(1 + math.exp(-z))
````

````python
#Given

x = 4
y = 1
w = 1.0
b = -3.0
alpha = 0.1
````

````python
# Forward pass
z = w * x + b
y_hat = sigmoid(z)
````

`dw=(y^​−y)x,db=(y^​−y)`

When we do backprop, we compute derivatives: 

````python
#Loss function
dz = y_hat - y

# Backpropagation
dw = dz * x
db = dz
````

````python
# Weight update (Gradient Descent)
w = w - alpha * dw
b = b - alpha * db
````

```python
| Part             | Where it appears        |
| ---------------- | ----------------------- |
| Loss function    | Inside `dz = y_hat - y` |
| Backpropagation  | `dw`, `db` calculation  |
| Gradient Descent | `w = w - α·dw`          |
```


````python
print("Updated weight:", w)
print("Updated bias:", b)
````
