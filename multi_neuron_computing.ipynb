```python
import math
```

```python
def sigmoid(z):
    return 1 / (1 + math.exp(-z))
```

```python
def log_loss(y, y_hat):
    epsilon = 1e-15
    y_hat = max(min(y_hat,1-epsilon),epsilon)
    return -(y * math.log(y_hat) + (1 - y) * math.log(1 - y_hat))
```

```python
# ---------- Data (single data point for clarity) ----------
x = 2
y = 1
```

```python
# ---------- Hyperparameters ----------
alpha = 0.1
epochs = 50
```

```python
# ---------- Initialize weights ----------
w1, b1 = 0.5, 0.0    # hidden neuron 1
w2, b2 = -0.5, 0.0   # hidden neuron 2
w3, w4, b3 = 1.0, 1.0, 0.0  # output neuron
```

```python
# ---------- Training loop ----------
for epoch in range(epochs):

    # ===== FORWARD PASS =====
    z1 = w1 * x + b1
    a1 = sigmoid(z1)

    z2 = w2 * x + b2
    a2 = sigmoid(z2)

    z3 = w3 * a1 + w4 * a2 + b3
    y_hat = sigmoid(z3)

    loss = log_loss(y, y_hat)

    # ===== BACKPROPAGATION =====
    dz3 = y_hat - y

    dw3 = dz3 * a1
    dw4 = dz3 * a2
    db3 = dz3

    dz1 = dz3 * w3 * a1 * (1 - a1)
    dz2 = dz3 * w4 * a2 * (1 - a2)

    dw1 = dz1 * x
    db1 = dz1

    dw2 = dz2 * x
    db2 = dz2

    # ===== GRADIENT DESCENT =====
    w3 -= alpha * dw3
    w4 -= alpha * dw4
    b3 -= alpha * db3

    w1 -= alpha * dw1
    b1 -= alpha * db1

    w2 -= alpha * dw2
    b2 -= alpha * db2

    # ===== MONITOR LEARNING =====
    if epoch % 5 == 0:
        print(f"Epoch {epoch:02d} | Loss: {loss:.4f} | Prediction: {y_hat:.4f}")
```
